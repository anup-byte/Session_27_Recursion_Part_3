```python
def fact(n):
    if n == 0:
        return 1
    smallOutput = fact(n-1)
    return n * smallOutput
```

```python
# Base case or Task 1 : l == 0 or l == 1 return True

#I. H : It works for l-1 sized lists

#Task2 or  I.Step : check [0] > [1]
                    return False
#            else:
 #       check for smaller list starting from [1.....end]
```

```python
def isSorted(a):
    l = len(a)
    if l == 0 or l == 1:
        return True
    
    if a[0] > a[1]:
        return False
    
    
    smallerList = a[1:]
    isSmallerListSorted = isSorted(smallerList)
    
    if isSmallerListSorted:
        return True
    
    else:
        return False
 
```

```python
a = [5,2,3,37,9,25]
```

```python
isSorted(a)
```

```python
isSorted(b)
```
